### FIlter data

> Advance Filtering with "**df.loc[]**" function

Syntax of "**df.loc[]**"
`df.loc[row(Label), columnLabel]`

[ **Note** ]: While invoking multiple row & column names, we've to enclose them using square-brackets.

+ Using boolean expression <br/>
    [ Note ]: <br/>
    + for using the Python's "and" inside of the pandas-dataframe, use "**&**";
    + for using the Python's "or" inside of the pandas-dataframe, use "**|**";
+ Using ranges of rows & columns labels
+ Using multiple rows & columns in a list-format
+ Using regex-filtration (filter based on textual pattern)

```python
# Display the totals of each type-1 of pokemons
data_csv["Type 1"].value_counts()
```

```python
# Only view the pokemons of type-1="Grass" values;
# Use boolean expression inside the "df.loc[]" function for filtration;
# Inside the "df.loc[]" function, we're checking the value of "Type 1=Grass" through every rows of the datafrmae;

# Single conditional expression
data_filter1 = data_csv[data_csv["Type 1"] == "Grass"]
data_filter1
```

```python
# [ Task ]: Display all the pokemons of "Type 1 == Grass" & "HP >= 50"
# Same boolean expression filtration;
# Multiple conditional expressions;
# Need to enclose each conditions inside paranthesis for seperating then;
# Use "**and(&)**" conditionals;
data_filter2 = data_csv[(data_csv["Type 1"] == "Grass") & (data_csv["HP"] >= 50)]
data_filter2
```

```python
# [ Task ]: Display all the pokemons of "Type 1 == Grass" or "Type 2 == Poison"
# Same boolean expression filtration;
# Multiple conditional expressions;
# Need to enclose each conditions inside paranthesis for seperating then;
# Use "**or(|)**" conditionals;
data_filter3 = data_csv[(data_csv["Type 1"] == "Grass") | (data_csv["Type 2"] == "Water")]
data_filter3
```

```python
# Task: Display only the pokemons of whether the "Type 1==Grass" or "Type 2==Water" and of "HP>70"
# Bothe use cases of "and(&)", "or(|)" in the same boolean filtration.
data_filter4 = data_csv[(data_csv["Type 1"] == "Grass") | (data_csv["Type 2"] == "Water") & (data_csv["HP"] >= 70)]
data_filter4
```

### Regex Filtering (Boolean Filtering on textual pattern)

> **pandas.Series.str.contains(self, pat, case=True, flags=0, na=None, regex=True)**

[**[Source] - Series.str.contains()**](https://github.com/pandas-dev/pandas/blob/v1.4.2/pandas/core/strings/accessor.py#L143-L3148)

[**[Doc] - pandas.Series.str()**](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.html)

+ Same "**df.loc[]**" function is called; boolean filtering is used while using the _regex-filtering_ on a particular column.

```python
# Display all the pokemons whose name contain "Mega".
data_csv.loc[data_csv["Name"].str.contains(pat="Mega")]
```

```python
# Display all the pokemons whose name do not contain "Mega".
data_csv.loc[~data_csv["Name"].str.contains(pat="Mega")]
```

```python
# Display all the pokemons whose "Type 1" of "Grass/Fire".
data_csv.loc[data_csv["Type 1"].str.contains(pat="Grass|Fire")] # the regex-attr is True by default
```

```python
# Display the pokemons of "Type 1" of both "Grass" or "Fire";
# For matching the string-literal regardless of the letter-casing, we need to define the "flags" attr to refer to the "re module".
import re
data_csv.loc[data_csv["Type 1"].str.contains(pat="grass|fire", flags=re.I)] # matches the string-literal pattern regardless of being case-sensitive
```

```python
# Display the pokemon names started with "pi".
# Symbol "^" => Caret/ Circumflex Accent

data_csv.loc[data_csv["Name"].str.contains(pat="^pi", flags=re.I)]
```

### df.groupby()

[**Docs**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html)

>Group DataFrame using a mapper or by a Series of columns. <br/>
A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

Pandas groupby is used for grouping the data according to the categories and apply a function to the categories. It also helps to aggregate data efficiently.

Pandas `dataframe.groupby()` function is used to split the data into groups based on some criteria.

**Returns "dataframe"**

```python
# Cluster the dataframe based on the categorical values of "Generation" column.
# data_csv.groupby("Generation") # Provide a "dataframe.groupby" object
for name, group in data_csv.groupby("Generation"):
    print("#"*50)
    print(name)
    print("#"*50)
    print(group)
```

```python
dir(data_csv.groupby("Generation"))
```

### Aggregated Statistics (Sum, Mean, Count)
**Use some math operations after the "groupby" function**
+ #### mean()
+ #### sum()
+ #### counting()


#### Execute operations (sum, mean & count) after grouping the dataframe based on the "Type 1" column.

```python
# Cluster/group the the pokemons based on the values of "Type 1" column.
data_csv.groupby("Type 1").mean()
```

```python
# DIsplay the aggregated count of each pokemon of "Type 1" column
data_csv["Type 1"].value_counts()
```

```python
# View the highest defense of pokemons using groupby on the "Type 1" column
# Clustered the pokemons based on the "Type 1" column
data_csv.groupby("Type 1").mean().sort_values("Defense", ascending=False)
```

```python
# Culster the pokemons based on the "Type 1" column;
# Then sore the pokemons of the hightest attack value using the "sort_values" function;
data_csv.groupby("Type 1").mean().sort_values("Attack", ascending=False)
```

```python
# Cluster/groupedby the pokemons based on the "Type 1" column;
# Get the aggreagated values (totals) of the each numeric columns uisng the "sum()" function after the "groupby()" func;
data_csv.groupby("Type 1").sum()
```

```python
# Organize the clustered dataframe based on the "Total" column in descending order
data_csv.groupby("Type 1").sum().sort_values("Total", ascending=False)
```

```python
# Display how many non-NaN values avaialable after clustering/grouping-by based on the "Type 1" column;
# The non-NaN values will be silently ignored for counting
data_csv.groupby("Type 1").count()
```

```python
data_csv["Count"] = 1 # Create a new column "Count"; store the totals of 

data_csv.groupby("Type 1").count() # simultaneously counting the values available in each columns & assigning them into the "Count" column
```

```python
# simultaneously counting the values available in each columns & assigning them into the "Count" column;
# later, display only the "Count" column as a Series
data_csv.groupby("Type 1").count()["Count"]
```
